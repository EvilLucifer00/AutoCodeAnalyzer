```python
# Restore legacy behavior via the classic bridge
from langchain_classic.memory import ConversationSummaryMemory
from langchain_classic.chains import ConversationalRetrievalChain
```

```python
# 1. Standard / Utility Imports
import os
from git import Repo

# 2. Splitters (Now in their own dedicated package)
from langchain_text_splitters import Language, RecursiveCharacterTextSplitter

# 3. Community Tools (Loaders, Vectorstores, Legacy Chains/Memory)
from langchain_community.document_loaders.generic import GenericLoader
from langchain_community.document_loaders.parsers import LanguageParser
from langchain_community.vectorstores import Chroma


# 4. Gemini Specific Drivers
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
```

```python
!mkdir test_repo
```

```python
repopath = "test_repo/"
repo = Repo.clone_from("https://github.com/entbappy/End-to-end-Medical-Chatbot-Generative-AI.git", to_path=repopath)
```

```python
loader = GenericLoader.from_filesystem(repopath,
                                       glob = "**/*",
                                       suffixes=[".py"],
                                       parser = LanguageParser(language = Language.PYTHON,parser_threshold = 500)
)
```

```python
documents = loader.load()
```

```python
documents[0]
```

```python
#context aware splitting
documents_splitter = RecursiveCharacterTextSplitter.from_language(language = Language.PYTHON,
                                                                  chunk_size = 500,
                                                                  chunk_overlap = 20)
```

```python
#chunks creation
text = documents_splitter.split_documents(documents)
```

```python
text
```

```python
#load the api key
from dotenv import load_dotenv
import os
load_dotenv()

# Assuming your .env file has GOOGLE_API_KEY=your_gemini_free_key
GOOGLE_API_KEY = os.environ.get("GOOGLE_API_KEY")
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
```

```python
# 1. Update to the new stable embedding model
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001") 

# 2. Point to a fresh vector database directory
vectordb = Chroma.from_documents(text, embedding=embeddings, persist_directory="./db")
```

```python
#initializing the LLM model
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.3)
```

```python
#creating the history memory to remmeber the previous context
memory = ConversationSummaryMemory(llm = llm, memory_key = "chat_history",return_messages = True)
```

```python
# creating the final chain
qa = ConversationalRetrievalChain.from_llm(
    llm, 
    retriever = vectordb.as_retriever(search_type = "mmr", search_kwargs = {"k": 8}),
    memory = memory 
)
```

```python
question = "what is download_hugging_face_embeddings function"
```

```python
result = qa.invoke({"question": question})

print(result['answer'])
```
